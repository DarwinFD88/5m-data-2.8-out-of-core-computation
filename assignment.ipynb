First, import polars as pl

```python
import polars as pl
```

Create 2 DataFrames

```python
metadata_pl = pl.read_csv("data/movies_metadata.csv",
                          infer_schema_length=100000)
```

```python
metadata_pl.head()
```

```python
#Convert 'id' column to Int64
metadata_pl = metadata_pl.with_columns(
    pl.col('id').cast(pl.Int64, strict=False)
)
```

```python
metadata_pl.head()
```

```python
ratings_pl = pl.read_csv("data/ratings.csv")
```

```python
ratings_pl.head()
```

Join the `metadata_pl` and `ratings_pl` DataFrames in Polars, then calculate the average rating by `genre` and by `original_language` (separately).

```python
#Join both DataFrames
joined_df = metadata_pl.join(ratings_pl, left_on='id', right_on='movieId', how='inner')
```

```python
#Average rating by genre
avg_ratings_by_genre = (joined_df.lazy()
                        .groupby('genres')
                        .agg(pl.col('rating').mean().alias('average_rating'))
                        .sort('average_rating', descending=True))
avg_ratings_by_genre.collect()
```

```python
#Average rating by original language
avg_ratings_by_original_language = (joined_df.lazy()
                                    .groupby('original_language')
                                    .agg(pl.col('rating').mean().alias('average_rating'))
                                    .sort('average_rating', descending=True))
avg_ratings_by_original_language.collect()
```

### Question 2
Question: Calculate the average total amount for vendors with at least 5 trips from the NYC Taxi Trip Data.

```python
q2 = (pl.scan_csv('data/taxi_trip_data.csv')
      .groupby('vendor_id')
      .agg([pl.count('total_amount').alias('total_trips'),
            pl.mean('total_amount').alias('average_amount')])
      .filter(pl.col('total_trips') >= 5)
      .sort('average_amount', descending=True)
      )

q2.collect(streaming=True)
```

```python
result = (
    pl.scan_csv('data/taxi_trip_data.csv')
    .groupby('vendor_id')
    .agg([
        pl.count('total_amount').alias('trip_count'),
        pl.mean('total_amount').alias('avg_total_amount')
    ])
    .filter(pl.col('trip_count') >= 5)
    .sort('avg_total_amount', descending=True)
    .collect(streaming=True)
)

result
```
